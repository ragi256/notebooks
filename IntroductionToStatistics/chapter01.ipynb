「統計データの分析プロセス」の節は重要。データ収集から始めるのではなく、何を対象にどんな分析をするか考えた上でデータを集める。
むやみにデータを集めて分析しても意味がない。

データを手に入れて分析作業を行う。パッケージが有るといっても適切な使い分けができないといけない。
結果を解釈し、適切に表現する手段を用いる。

下記の4プロセスを行い、結果をフィードバックさせて繰り返し分析をする。
1. 仮設を立てて分析対象を明らかにする
2. データを収集する
3. 適切な手段を用いて分析する
4. 結果を適切に表現する

統計が「ウソつき」の場合上記プロセスのうち1個か2個Trickが入る。

## 年齢、出生、脂肪、配偶関係などに次のような誤りが生じる状況、理由を説明せよ

2. 燃焼女子の年齢の歌唱申告（20＋$x$歳のものが20歳と申告するなど）
3. 高年齢者の年齢の過大申告
4. 4月の出生が多く申告される
5. 有配偶者が男女等しくなく、婚姻統計では女の有配偶者数が男のそれを上回る
